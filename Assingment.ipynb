```python
import pandas as pd
import csv
def main():
    df_quant=pd.DataFrame(pd.read_csv("Input_StartOfDay_Positions.csv"))
    df_trans=pd.DataFrame(pd.read_json("1537277231233_Input_Transactions_j.json"))
    quantityE=0
    quantityI=0
    dict_B={}
    dict_S={}
    dict_E={}
    dict_I={}
    ins_list=[]
    for index, row in df_trans.iterrows():
        #print("transaction type",row["TransactionType"])
        trans_type=row["TransactionType"]
        ins_type=row['Instrument']
        #print("The iNstrument List is",ins_list)
        if row["TransactionType"] == "B":
            #print("The INSTRUMENT IS ",row['Instrument'])
            if row['Instrument'] in ins_list:
                quantity_e=dict_E.get("E:"+ins_type+":101")

            else:
                temp_e=df_quant.loc[df_quant['AccountType'].isin(['E']) & df_quant['Instrument'].isin([row['Instrument']])]['Quantity']
                quantity_e=temp_e.values[0]
            quantityE=quantity_e+float(row["TransactionQuantity"])
            trans_type="E"
            dict_E["E:"+ins_type+":101"]=quantityE
            ##print("the E quant is",quantityE)
            if row['Instrument'] in ins_list: 
                quantity_i=dict_I.get("I:"+ins_type+":201")
            else:
                temp_i=df_quant.loc[df_quant['AccountType'].isin(['I']) & df_quant['Instrument'].isin([row['Instrument']])]['Quantity']
                quantity_i=temp_i.values[0]
            quantityI=quantity_i-float(row["TransactionQuantity"])
            #print("the I quant is",quantityI)
            trans_type="I"
            dict_I["I:"+ins_type+":201"]=quantityI
        if row["TransactionType"] == "S":
            #print("The INSTRUMENT IS ",row['Instrument'])
            if row['Instrument'] in ins_list:
                quantityE=dict_E.get("E:"+ins_type+":101")
                quantityI=dict_I.get("I:"+ins_type+":201")
            else:
                temp_e=df_quant.loc[df_quant['AccountType'].isin(['E']) & df_quant['Instrument'].isin([row['Instrument']])]['Quantity']
                quantityE=temp_e.values[0]
                temp_i=df_quant.loc[df_quant['AccountType'].isin(['I']) & df_quant['Instrument'].isin([row['Instrument']])]['Quantity']
                quantityI=temp_i.values[0]
            quantityE=quantityE-float(row["TransactionQuantity"])
            trans_type="E"
            dict_E["E:"+ins_type+":101"]=quantityE
            #print("the E quant is",quantityE)
            quantityI=quantityI+float(row["TransactionQuantity"])
            dict_I["I:"+ins_type+":201"]=quantityI
            #print("the I quant is",quantityI)
        if row['Instrument'] not in ins_list:
            ins_list.append(row['Instrument'])

    print("The E Dict is ",dict_E)
    print("The I Dict is ",dict_I)

    myFile = open('final_output.csv', 'w')
    with myFile:
        myFields = ['Instrument', 'Account','AccountType','Quantity']
        writer = csv.DictWriter(myFile, fieldnames=myFields)    
        writer.writeheader()
        for keys, values in dict_E.items():
            key_split=keys.split(":")
            account_type=key_split[0]
            instrument=key_split[1]
            account=key_split[2]
            writer.writerow({'Instrument' : instrument,'Account':account,'AccountType':account_type,'Quantity':values})
        for keys, values in dict_I.items():
            key_split=keys.split(":")
            account_type=key_split[0]
            instrument=key_split[1]
            account=key_split[2]
            writer.writerow({'Instrument' : instrument,'Account':account,'AccountType':account_type,'Quantity':values})

    
```

```python
main()
    
    

        

        
        
```
